`-` 필수: 파이썬입문 (수업시간에 별도로 파이썬을 리뷰하는 시간을 갖지

`-` 딥러닝의 기초: DNN, 손실함수, 옵티마이저, 역전파, universal
approximation thm

`-` 이미지 자료 분석: CNN, Class Activation Map (CAM) and XAI

`-` 추천시스템: SVD, Collaborative Filtering …

`-` 텍스트와 시퀀스 자료 분석: RNN, LSTM, GRU, Attention…

`-` 생성모형: 식별모형과 생성모형, GAN
